```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv')
df
```

##1.Create a bar graph of the maximum ‘price’ by ‘neighbourhood_group’. Include the ‘price’ in your graph

```python
df.groupby('neighbourhood_group').price.agg('max').plot(kind = 'bar',
                                                 title = 'Maximum Cost by Neighborhood Group',
                                                 ylabel= 'Price',
                                                 xlabel= 'Neighborhood')
```

Here you can see clearly that the Bronx are the cheapest neighborhood and Brooklyn, Manhattan, and Queens all have the same maximum cost then Staten Island is somewhere in the middle. Also I hate British people.

##2. Create a multiple bar graph with ‘neighbourhood_group’ and ‘room_type’ by looking at the average ‘price’.

```python
df.groupby(['neighbourhood_group','room_type']).price.agg('mean').unstack().plot(kind = 'bar',
                                                                                 xlabel='Neighborhood',
                                                                                 ylabel='Price')
```

I wanted my bars to look pretty like the ones that were on your github so I found that this .unstack feature made it to where the rooms weren't individually spread out and instead groupped by what neighborhood they reside in.
